```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Using NumPy to generate random data
np.random.seed(42)  # For reproducibility
data = np.random.randn(100, 3)  # 100 rows, 3 columns of random data
```

```python
# Step 2: Using Pandas to create a DataFrame
df = pd.DataFrame(data, columns=['A', 'B', 'C'])
```

```python
# Display the first few rows of the DataFrame
print("DataFrame:\n", df.head())
```

```python
# Step 3: Using Matplotlib to visualize the data
plt.figure(figsize=(8, 6))
```

```python
# Plot all three columns
plt.plot(df['A'], label='Column A')
plt.plot(df['B'], label='Column B')
plt.plot(df['C'], label='Column C')
```

```python
# Add title and labels
plt.title('Random Data Visualization')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
```

```python
# Show the plot
plt.show()
```
